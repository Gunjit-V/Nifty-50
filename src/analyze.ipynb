```python
import pandas as pd
import plotly.graph_objects as go
import nbformat
```

```python
df = pd.read_parquet("Nifty 50_ohlc_2025-10-06.parquet")
df = df[::-1].reset_index(drop=True)
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['time'], open=df['into'], high=df['inth'], low=df['intl'], close=df['intc'])])
fig.update_layout(title='Nifty 50 OHLC Data', xaxis_title='Time', yaxis_title='Price')
fig.show()
```
